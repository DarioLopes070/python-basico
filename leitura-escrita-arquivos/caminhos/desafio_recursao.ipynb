### Desafio para resolução:



criar um código que irá percorrer todos os arquivos de determinado diretório e retornar o tamanho deles e das pastas, de acordo com a profundidade desejada, então o código deve receber 2 parametros, o primeiro sendo a o diretório em si e o segundo a profundidade de pesquisa, então deve retornar nome e tamanho.

```python
import os
from pathlib import Path
arquivos_extraidos = []
tamanho_arquivos = []
def pegar_arquivos_pasta(pasta):
    lista_arquivos = os.listdir(pasta)
    # lista_arquivos = pasta.glob("*")
    for arquivo in lista_arquivos:
        arquivo = Path(Path.cwd()/pasta/arquivo)
        if arquivo.is_file():
            nome_arquivo = arquivo.name
            tamanho_arquivo = os.path.getsize(arquivo)
            arquivos_extraidos.append(nome_arquivo) 
            tamanho_arquivos.append(tamanho_arquivo)
        elif arquivo.is_dir():
            pegar_arquivos_pasta(Path(Path.cwd()/pasta/arquivo))
```

```python
pegar_arquivos_pasta(Path("D:\\dario") / "Documents")  
x=0
for arquivo in arquivos_extraidos:
    print(f"\nArquivo: {arquivo} ------ {tamanho_arquivos[x]}") 
    x+=1
print(f"Soma total de tamanho em GB: {(sum(tamanho_arquivos))/(1024*1024*1024)}")   
```

```python
# Código para montar a hierarquia de pastas e arquivos
import os
from pathlib import Path

# hierarquia = {}
def montar_arvore(pasta, hierarquia):
    lista_arquivos = os.listdir(pasta) 
    if not hierarquia:
        hierarquia.setdefault(pasta.name, {})
    for arquivo in lista_arquivos:
        arquivo = Path(Path.cwd()/pasta/arquivo)
        nome_pasta = pasta.name
        if arquivo.is_file():
            nome_arquivo = arquivo.name
            tamanho_arquivo = os.path.getsize(arquivo)
            hierarquia.setdefault(nome_pasta, {}).setdefault(nome_arquivo, {}).setdefault("tamanho", tamanho_arquivo)
            hierarquia.setdefault(nome_pasta, {}).setdefault(nome_arquivo, {}).setdefault("tipo", "arquivo")
        elif arquivo.is_dir():
            montar_arvore(Path.cwd()/pasta/arquivo, hierarquia.setdefault(nome_pasta, {}))
                    
# montar_arvore(Path("D:\\dario") / "Documents", hierarquia)  
# print(hierarquia)
```

```python
# Código para Calcular tamanho das de pastas
def calcular_tamanho(hierarquia):
    soma_tamanho = 0
    
    for nome, conteudo in hierarquia.items():
        if nome == "tamanho":
            continue
        if conteudo.get("tipo") == "arquivo":
            soma_tamanho += hierarquia[nome]["tamanho"]
        else:
            soma_tamanho += calcular_tamanho(hierarquia[nome])
    hierarquia.setdefault("tamanho", soma_tamanho)
    return soma_tamanho  
# raiz = next(iter(hierarquia))              
# # raiz = list(hierarquia.keys())[0]             
# calcular_tamanho(hierarquia[raiz])  
# pprint(hierarquia)
```

```python
# Código para Calcular tamanho das de pastas
def exibir_arvore(hierarquia, nivel=0, profundidade=0, prefixo=""):
    exibir = ""
    if profundidade < nivel:
        return ""
    for nome, conteudo in hierarquia.items():
        if nome == "tamanho" or nome == "tipo":
            continue
        if conteudo.get("tipo") == "arquivo":
            icone = "📄"
        else:
            icone = "📁"
        ultimo_item = [
            chave for chave in hierarquia.keys() if chave not in ["tamanho", "tipo"]
        ][-1]
        eh_ultimo = nome == ultimo_item
        if eh_ultimo:
            conector = "└── "
        else:
            conector = "├── "
        if eh_ultimo:
            novo_prefixo = prefixo + "    "
        else:
            novo_prefixo = prefixo + "│   "

        tamanho = (
            (conteudo.get("tamanho")) / (1024**2) if conteudo.get("tamanho") else 0
        )
        texto_visual = prefixo + conector + icone + " " + nome
        exibir += f"{texto_visual:<100}{tamanho:.2f}MB\n"
        exibir += exibir_arvore(hierarquia[nome], nivel + 1, profundidade, novo_prefixo)
    return exibir


# exibir = exibir_arvore(hierarquia, profundidade=1)
# print(exibir)
```
